```python
xray_folder=os.listdir('xray_train')
xray_folder
```

```python
train_images=[]
labels=[]
folders=['normal', 'pneumonia']
```

```python
for folder in xray_folder:
    print(folder)
```

```python
for folder in xray_folder:
    xray_train=f'.\\xray_train\\{folder}'
    print(os.listdir(xray_train))
```

```python
class preprocess:
    def __init__(self, Path: str)->None:
        self.path=Path
        
    def process_and_normalized(self):
        image=io.imread(self.path)
        resized_img=transform.resize(image, output_shape=(150,150))
        return resized_img.flatten()
```

```python
for folder in xray_folder:
    xray_train=f'.\\xray_train\\{folder}'
    for file in (os.listdir(xray_train)):
        from_path=xray_train+'\\'+file
        processor=preprocess(from_path)
        vector=processor.process_and_normalized()
        train_images.append(vector)
        labels.append(folders.index(folder))
        
```
